```python
gotnya_analys = pd.read_csv(r"C:\Users\66910\Downloads\BGD_Winter_Explore\step 4\3_heatig_period.csv")
```

```python
gotnya_analys.reset_index(inplace=True)
gotnya_analys.columns = ["start","end","long","minday", "minfday", "meanperiod", "gsop","del"]
gotnya_analys
```

```python
gotnya_analys = gotnya_analys[["start","end","long","minday", "minfday", "meanperiod", "gsop"]]
gotnya_analys
```

```python
# удаляем ненужные символы
gotnya_analys['start'] = gotnya_analys['start'].str.replace("(", '')
gotnya_analys['start'] = gotnya_analys['start'].str.replace(")", '')
gotnya_analys['start'] = gotnya_analys['start'].str.replace("'", '')
gotnya_analys['start'] = gotnya_analys['start'].str.replace("[", '')
gotnya_analys['start'] = gotnya_analys['start'].str.replace("/", '-')
gotnya_analys['gsop'] = gotnya_analys['gsop'].str.replace(")", '')
gotnya_analys['gsop'] = gotnya_analys['gsop'].str.replace("]", '')
gotnya_analys['end'] = gotnya_analys['end'].str.replace("(", '')
gotnya_analys['end'] = gotnya_analys['end'].str.replace(")", '')
gotnya_analys['end'] = gotnya_analys['end'].str.replace("'", '')
gotnya_analys['end'] = gotnya_analys['end'].str.replace("[", '')
gotnya_analys['end'] = gotnya_analys['end'].str.replace("/", '-')
gotnya_analys
```

```python
gotnya_analys["start"] = gotnya_analys["start"].str.strip()
gotnya_analys["end"] = gotnya_analys["end"].str.strip()
gotnya_analys["start"] = pd.to_datetime(gotnya_analys["start"],format='%m-%d-%Y')
gotnya_analys["end"] = pd.to_datetime(gotnya_analys["end"],format='%m-%d-%Y')
```
