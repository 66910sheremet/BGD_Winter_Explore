```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
full_data = pd.read_excel(r"C:\Users\66910\Downloads\BGD_Winter_Explore\Step 5\1_full_data_gotnya_2023.xlsx")
full_data
```

```python
full_data['MA'] = full_data['mean_year'].rolling(window=10).mean()
full_data['MAlong'] = full_data['long'].rolling(window=10).mean()
full_data['MAminday'] = full_data['minday'].rolling(window=10).mean()
full_data['MAminfday'] = full_data['minfday'].rolling(window=10).mean()
full_data['MAmeanperiod'] = full_data['meanperiod'].rolling(window=10).mean()
full_data['MAgsop'] = full_data['gsop'].rolling(window=10).mean()
```

```python
full_data.rename(columns=lambda x: x.strip())
```

```python
full_data.set_index('year', inplace=True)
```

```python
MA = full_data[['mean_year','MA']]
MAlong = full_data[['long','MAlong']]
MAminday = full_data[['minday','MAminday']]
MAminfday = full_data[['minfday','MAminfday']]
MAmeanperiod = full_data[['meanperiod','MAmeanperiod']]
MAgsop = full_data[['gsop','MAgsop']]
```

```python
plt.figure(figsize=(9, 5));
MA.plot();
plt.ylabel('Температура в градусах Цельсия')
plt.xlabel('Год измерения')
plt.grid(True)
plt.title("Среднегодовые минимальные округленные температуры Белгорода ")
plt.rcParams['font.family'] = 'serif'
plt.rcParams.update({'font.size': 12})
plt.show();
```

```python
plt.figure(figsize=(9, 5));
MAlong.plot();
plt.ylabel('Продолжительность ОП, дней')
plt.xlabel('Год измерения')
plt.grid(True)
plt.title("Продолжительность отопительного периода Готни")
plt.rcParams['font.family'] = 'serif'
plt.rcParams.update({'font.size': 12})
plt.show();
```

```python
plt.figure(figsize=(9, 5));
MAminday.plot();
plt.ylabel('Температура в градусах Цельсия')
plt.xlabel('Год измерения')
plt.grid(True)
plt.title("Температура наиболее холодного \n дня отопительного периода Готни")
plt.rcParams['font.family'] = 'serif'
plt.rcParams.update({'font.size': 12})
plt.show();
```

```python
plt.figure(figsize=(9, 5));
MAminfday.plot();
plt.ylabel('Температура в градусах Цельсия')
plt.xlabel('Год измерения')
plt.grid(True)
plt.title("Температуры наиболее холодной \n пятидневки Готни")
plt.rcParams['font.family'] = 'serif'
plt.rcParams.update({'font.size': 12})
plt.show();
```

```python
plt.figure(figsize=(9, 5));
MAmeanperiod.plot();
plt.ylabel('Температура в градусах Цельсия')
plt.xlabel('Год измерения')
plt.grid(True)
plt.title("Средние температуры отопительного \n периода Готни")
plt.rcParams['font.family'] = 'serif'
plt.rcParams.update({'font.size': 12})
plt.show();
```

```python
plt.figure(figsize=(9, 5));
MAgsop.plot();
plt.ylabel('Градусосутки отопительного периода')
plt.xlabel('Год измерения')
plt.grid(True)
plt.title("Градусосутки отопительного периода Готни")
plt.rcParams['font.family'] = 'serif'
plt.rcParams.update({'font.size': 12})
plt.show();
```
