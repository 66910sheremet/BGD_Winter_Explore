```python
# добавили колонку с датой путем конкотенации
df['time_explore'] = df['Year'].astype(str) + '-' + df['month'].astype(str) + '-' + df['day'].astype(str)
df['time_explore']= pd.to_datetime(df['time_explore'])
df
```

```python
# установали в качестве индекса
df_base = df.set_index('time_explore')
```

```python
df_base
# Заготовка из которой можно вытягивать нужную информацию!
```

# Работаем со средними температурами

```python
df_meanT = df_base[['MeanT']]
df_meanT
# Потянули среднедневную температуру
```

```python
df_meanT = df_meanT.loc[~((df_meanT.index.month==2) & (df_meanT.index.day==29))]
# удалили 29 февраля из всех годов!
```

```python
ser = df_meanT.squeeze()
# Перевели в Series
```

```python
# сгруппировали по году данные
groups = ser.groupby(pd.Grouper(freq='A'))
```

```python
# создали пустой датафрейм
years_mean_T_365 = pd.DataFrame()
```

```python
for name, group in groups:
  years_mean_T_365[name.year] = group.values
```

```python
years_mean_T_365
```

```python
years_mean_T_365.drop([1941,1942,1943, 1944,1945,1966,1976 ], axis= 1 , inplace= True )
# удалили года с пропущенными значениями
```

```python
# вытянуди среднегодовые температруры
mean_year_mean = years_mean_T_365.mean()
mean_year_mean
```

```python
years_mean_T = mean_year_mean.to_frame (name='values')
# Переводим Series во Frame
```

```python
years_mean_T['valuesMA'] = years_mean_T.rolling(window=10).mean()
# Строим средние скользящие
```

```python
plt.figure(figsize=(9, 5))
years_mean_T.plot()
plt.ylabel('Температура в градусах Цельсия')
plt.xlabel('Год измерения')
plt.grid(True)
plt.title("Среднегодовые средние округленные температуры Белгорода ")
plt.rcParams['font.family'] = 'serif'
plt.rcParams.update({'font.size': 12})
plt.show()
```
